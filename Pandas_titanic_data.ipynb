## data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#### drawaing histogram hist
```

```python
plt.hist(x=df['Age'],)
```

```python
plt.hist(x=df['Age'],color='yellow')
```

```python
plt.hist(x=df['Age'],color='yellow')
plt.title('Distribution of Age',fontsize =20,color='blue')
plt.xlabel('Age',fontsize=15,color='red')
plt.ylabel('Frequency',fontsize=15,color='red')
plt.show()
```

```python
sns.boxplot(x=df["Age"])
plt.title('Destribution of Age')
plt.xlabel("age")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Gender Survey",fontsize=20,color="red")
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="barh")
plt.title("Gender Survey",fontsize=20,color="red")
plt.xlabel('Count')
plt.ylabel('Sex')
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='bar',color='teal')
plt.title("Pclass Destribution")
plt.ylabel("No. of Passenger")
plt.xlabel('Class')
plt.show()
```

```python
emb=df['Embarked'].value_counts()
keys=emb.keys().to_list()
counts=emb.tolist()
plt.pie(x=counts,labels=keys,autopct="%1.1f%%")
plt.show()
```

```python
#### imporvise
```

```python
from matplotlib import as cm
emb=df['Embarked'].value_counts()
keys=emb.keys().to_list()
counts=emb.to_list()
cs=cm.set1([2,4,6,8])
plt.pie(x=counts,labels=keys,autopct="%1.1f%%",colors=cs)
plt.show()
```
